```r
# Generate random data
x <- rnorm(100)
y <- rnorm(100)

# make background white so it's visible with dark theme
par(bg = "white")

# Create scatter plot
plot(x, y, main = "Scatter Plot", xlab = "X", ylab = "Y")

# Add a regression line
abline(lm(y ~ x), col = "red")

# Add a legend
legend("topleft", legend = "Regression Line", col = "red", lty = 1, cex = 0.8)

# Make stacked bar chart
data <- matrix(c(10, 20, 30, 40, 50, 60), nrow = 2, byrow = TRUE)
barplot(data,
    main = "Stacked Bar Chart", xlab = "Category",
    ylab = "Value", col = c("blue", "red"), legend = c("Group 1", "Group 2")
)
```
